Seaborn — популярная библиотека готовых шаблонов для статистической визуализации, написанная на бэкенде matplotlib. Преимущества seaborn:

* Выразительный высокоуровневый интерфейс: построение большинства простых графиков происходит в одну строчку кода.
* Более эстетичные графики: часто встроенные в seaborn стили весьма приятны и хороши уже "из коробки".
Автор seaborn — Michael Waskom, PhD, сотрудник Center for Neural Research (Нью-Йорк), выпускник Стэнфорда. Разработка seaborn это его хобби, которое делает жизни тысяч людей чуть лучше.

```python
import seaborn as sns
import matplotlib.pyplot as plt
penguins = sns.load_dataset("penguins")
```

```python
sns.set_style("darkgrid")
```

```python
penguins
```

```python
plt.figure(figsize=(13,13))
sns.histplot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack")
```

```python
plt.figure(figsize=(13,13))
sns.kdeplot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack")
```

```python
sns.displot(data=penguins, x="flipper_length_mm", hue="species", col="species")
```

```python
sns.displot(data=penguins, x="flipper_length_mm", hue="species", col="species", kind='kde')
```

```python
plt.figure(figsize=(11,9))
sns.scatterplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue="species")
```

```python
plt.figure(figsize=(11,9))
sns.scatterplot(data=penguins, x="species", y="island")
```

```python
plt.figure(figsize=(11,9))
sns.histplot(data=penguins, x="species", hue="species", shrink=.3, alpha=.5, legend=False)
```

```python
sns.jointplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue="species")
```

```python
sns.pairplot(data=penguins, hue="species")
```

## Реализуем один из наших кейсов при помощи seaborn!

```python
import pandas as pd
entrants = pd.read_csv("entrants.csv", names = ['#', 'Согласие', 'Согласие в университете', 'Сумма баллов', 'Экзамен 1 (математика)','Экзамен 2 (математика)', 'Экзамен 3 (физика)', 'Экзамен 4 (Русский Язык)', 'ИД - аттестат', 'ИД - итоговое сочинение', 'ИД - ГТО', 'Преимущество', 'Общежитие', 'Статус'])
entrants.rename(columns = {'Сумма баллов':'сумма',
                           'Экзамен 1 (математика)':'М1',
                           'Экзамен 2 (математика)':'М2',
                           'Экзамен 3 (физика)':'Ф',
                           'Экзамен 4 (Русский Язык)':'РЯ',
                           'ИД - аттестат':'аттестат',
                           'ИД - сочинение':'сочинение',
                           'ИД - ГТО':'ГТО',
                           'Преимущество':'преимущество',
                           'Общежитие':'общежитие'},
                inplace = True)
entrants["Согласие"] = (entrants["Согласие"] == "Да")
entrants["Согласие в университете"] = (entrants["Согласие в университете"] == "Да")
entrants["преимущество"] = (entrants["преимущество"] == "Да")
entrants["общежитие"] = (entrants["общежитие"] == "Да")
entrants.drop('Статус', axis=1, inplace=True)
entrants.drop('преимущество', axis=1, inplace=True)
```

```python
plt.figure(figsize=(13,13))
sns.histplot(data=entrants, x="сумма", hue="ГТО", multiple="stack", kde=True)
```

```python
plt.figure(figsize=(11,9))
sns.scatterplot(data=entrants, x="М1", y="М2")
```

```python
sns.jointplot(data=entrants, x="М1", y="М2")
```

```python
plt.figure(figsize=(11,9))
sns.barplot(data=entrants,x='М1',y='М2')
```

```python
plt.figure(figsize=(11,9))
sns.boxplot(data=entrants,x='М1',y='М2')
```

```python
sns.jointplot(data=entrants, x="Ф", y="М2")
```

```python
plt.figure(figsize=(11,9))
sns.boxplot(data=entrants,x='Ф',y='М2')
```

```python
results = entrants[['Ф', 'М1', 'М2', 'РЯ', 'ГТО']]
sns.pairplot(data=results, hue="ГТО")
```
