```python
import numpy as np

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--')
plt.plot( t, -100*t**2 - t + 10, 'bs')
plt.plot( t, t**3, 'g^')

plt.show()
```

Несколько графиков

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.grid(True)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(figsize = (12,12))
plt.subplot(211)
plt.grid(True)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

# Пример: последовательность механических измерений при нагружении стального стержня

```python
import pandas as pd
data = pd.read_excel('Data.xlsx', names=['time', 'chanel 0', 'chanel 1', 'chanel 2'])
```

```python
data
```

```python
data = data.rename(columns = {data.columns[0]:'time', data.columns[1] : 'chanel 0', data.columns[2] : 'chanel 1', data.columns[3] : 'chanel 2'})
```

```python
data
```

```python
plt.figure(figsize = (12,12))
plt.xlabel('Время')
plt.ylabel('Перемещение, канал 0')
plt.grid()
plt.plot(data.time, data['chanel 0'])
```

```python
plt.figure(figsize = (12,12))
plt.xlabel('Время')
plt.ylabel('Перемещение, канал 1')
plt.grid()
plt.plot(data.time, data['chanel 1'])
```

Пока ничего не понятно. Посмотрим на все графики, изображенные на одном figure

```python
plt.figure(figsize = (20,15))
plt.subplot(311)
plt.xlabel('Время')
plt.ylabel('Перемещение, канал 0')
plt.grid()
plt.plot(data.time, data['chanel 0'])
plt.show()

plt.figure(figsize = (20,15))
plt.subplot(312)
plt.xlabel('Время')
plt.ylabel('Перемещение, канал 1')
plt.grid()
plt.plot(data.time, data['chanel 1'])
plt.show()

plt.figure(figsize = (20,15))
plt.subplot(313)
plt.xlabel('Время')
plt.ylabel('Нагрузка')
plt.grid()
plt.plot(data.time, data['chanel 2'])
plt.show()
```
