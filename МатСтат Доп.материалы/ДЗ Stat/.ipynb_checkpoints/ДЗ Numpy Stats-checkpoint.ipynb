# Задачи:
## 1	Что представляет собой библиотека SciPy?
## 2	В чём отличия и сходства NumPy и SciPy?
## 3	Перечислите основной набор пакетов для научных вычислений.

1. SciPy – это бесплатная программная библиотека с открытым исходным кодом, используемая для научных вычислений на языке Python. SciPy используется в инженерии, естественных науках, а также математических вычислениях. SciPy подходит для таких задач, как анализ данных, визуализация данных и численные вычисления.
2. NumPy и SciPy — это две важные библиотеки Python для численных вычислений, но они служат разным целям.
NumPy фокусируется на быстрой обработке и подходит для основных концепций массивов. Он в основном работает с многомерными объектами массивов, эффективно отслеживая тип данных, форму и размер.
SciPy предлагает детальные версии функций для глубоких численных вычислений. Он построен на основе NumPy и предлагает широкий спектр инструментов для сложных задач, таких как решение дифференциальных уравнений, оптимизация, статистический анализ и специализированные математические функции.
Выбор между NumPy и SciPy зависит от поставленной задачи. NumPy предпочтительнее для базовых вычислительных операций, манипуляций с массивами и элементарных математических вычислений, а SciPy подходит для продвинутых вычислительных процедур, статистического тестирования и разнообразных функций, таких как обработка изображений и градиентная оптимизация.
3. MATLAB, GNU Octave и SciLab "Наиболее распространённый среди научных работников пакет программ носит название MATLAB. Также существуют бесплатные пакеты программ, которые решают практически те же задачи: GNU Octave и SciLab. Кроме того, если нужно получить решение математической задачи в алгебраическом виде, учёные используют Mathematica, MathCAD, Maple."
