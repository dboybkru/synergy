# I. Scatterplots

```python
import numpy as np
import matplotlib.pyplot as plt

#Инструмент scatter
data = {'a': np.arange(50),#х-координаты
        'c': np.random.randint(0, 50, 50),#цвет
        'd': np.random.randn(50)*1}#диаметр
data['b'] = data['a'] + 20 * np.random.randn(50) #у-координаты
data['d'] = np.abs(data['d']) * 100

plt.figure(figsize=(12,12))
plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

## Часто используется для визуализаций наборов измерений величин, для которых не важна последовательность. Также для по scatterplot можно отследить корреляции величин

### Пример:
Мальчик Паша каждый день, приходя из школы, кормит свою собаку. Паша знает, что собаке нужно дать  200 грамм корма, однако, все зависит от пашиного настроения: если Паша получил хорошую оценку в школе, то настроение Паши - хорошее, и он может дать собаке немногоо больше корма. А если наоборот - Паша получил плохую оценку, то и корма будет меньше. Настроение собаки линейно зависит от количества корма, а настроение Паши - зависит от полученных Пашей оценок.

```python
Pashas_mood = np.random.randn(100)
Dogs_mood = 200 + Pashas_mood*50


plt.figure(figsize=(9,7))
plt.grid()
plt.scatter(Pashas_mood, Dogs_mood)
plt.xlabel('Настроение Паши')
plt.ylabel('Настроение Сообаки')
```

```python
"""
А как настроение собаки зависит от Пашиных оценок?
"""

Pashas_marks = np.random.choice([2,3,4,5],100)
Pashas_mood = (Pashas_marks - 3.5)/1.5 +  np.random.randn(100)/3

Dogs_mood = 200 + Pashas_mood*50

plt.figure(figsize=(9,7))
plt.grid()
plt.scatter(Pashas_marks, Dogs_mood, c='#ff8243')
plt.xlabel('Оценки Паши')
plt.ylabel('Настроение Сообаки')
```

```python
"""
А что, если настроение собаки тоже имеет свое отклонение?
"""
Pashas_mood = np.random.randn(100)
Dogs_mood = 200 + Pashas_mood*50 + np.random.randn(100)*10


plt.figure(figsize=(9,7))
plt.grid()
plt.scatter(Pashas_mood, Dogs_mood)
plt.xlabel('Настроение Паши')
plt.ylabel('Настроение Сообаки')
```

```python
x = np.array([4, 5, 6, 3, 6, 5, 7, 3, 4, 5])
y = np.array([3, 4, 1, 3, 2, 3, 3, 1, 2, 3])
z = np.array([6, 9, 8, 7, 9, 8, 9, 6, 8, 7])

x_mean = np.mean(x)
y_mean = np.mean(y)
z_mean = np.mean(z)

x_deviation = np.std(x)
y_deviation = np.std(y)
z_deviation = np.std(z)

bars = [x_mean, y_mean, z_mean]
bar_categories = ['X', 'Y', 'Z']
error_bars = [x_deviation, y_deviation, z_deviation]

plt.bar(bar_categories, bars, yerr=error_bars)
plt.show()
```

```python
def draw_custom_violin(df, title=''):
    plt.figure(figsize=(30,8))
    plt.title(title)
    df_cols = df.columns.values
    plt.xticks(1+np.arange(len(df_cols)), df_cols, rotation='vertical')
    plt.violinplot(df, showmedians=True);
```

```python
import pandas as pd
df = pd.DataFrame({'feature_1' : np.random.randn(1000)*10 + 5, 'feature_2' : np.random.randn(1000)*3 - 5, 'feature_3' : np.random.randn(1000)*4, 'feature_4' : np.random.randn(1000)*3 -+10})
```

```python
draw_custom_violin(df, 'Random DF')
```

```python
import pandas as pd
df = pd.DataFrame({'Треугольное' : np.random.triangular(-3, -1 , 15, size=100000), 
                   'Нормальное' : np.random.randn(100000)*3 - 5, 
                   'Экспоненциальное' : np.random.exponential(1, size=100000), 
                   'Равномерное' : np.random.uniform(-20,1,100000)})
```

## V. Линии уровня функций

```python
import numpy as np
x, y = np.mgrid[-3*np.pi:3*np.pi:100j,
                -3*np.pi:3*np.pi:100j]
z = np.sinc(x) + np.cos(y)

fig, ax = plt.subplots()

ax.contour(z)

fig.set_figwidth(8)    #  ширина и
fig.set_figheight(8)    #  высота "Figure"

plt.show()
```

```python
import numpy as np
x, y = np.mgrid[0:1:100j,
                0:1:100j]

z = 2*x*y/(x+y + 1e-3)


fig, ax = plt.subplots()

ax.contour(z, levels=30)

fig.set_figwidth(8)    #  ширина и
fig.set_figheight(8)    #  высота "Figure"

plt.show()
```

```python
#Число контуров увеличивается параметром levels

x, y = np.mgrid[-3*np.pi:3*np.pi:300j,
                -3*np.pi:3*np.pi:300j]
z = np.sinc(x) + np.cos(y)

fig, ax = plt.subplots()

ax.contour(z, levels = 50)

fig.set_figwidth(12)    #  ширина и
fig.set_figheight(12)    #  высота "Figure"

plt.show()
```

```python
#С помощью параметра colors мы можем задать один цвет для всех линий или, что очень важно для физиков и инженеров, 
#цвет каждой линии. В случае, если необходимо определить цвет каждой линии, то параметру colors необходимо присвоить 
#массив длинна которого соответствует количеству линий (levels) а каждый элемент определенному цвету


x, y = np.mgrid[0:10:100j, 0:10:100j]

#  Вычисляем напряженность электрического
#  поля двух зарядов в каждой точке плоскости:
x_1, x_2, y_0 = 4, 6, 5
d = x_2 - x_1
r_1 = np.sqrt((x - x_1)**2 + (y - y_0)**2)
r_2 = np.sqrt((x - x_2)**2 + (y - y_0)**2)
a = (d**2 - r_1**2 - r_2**2)/(2*r_1*r_2)
E_1 = 1/(r_1**2)
E_2 = 1/(r_2**2)
E = 10*np.sqrt(E_1**2 + E_2**2 + 2*E_1*E_2*np.cos(a))

fig, axes = plt.subplots(1, 2)

#  Задаем значение каждого уровня:
lev = [1, 2, 3, 4, 6, 10, 20, 40, 100, 900]

#  Контуры одного цвета:
axes[0].contour(x, y, E, levels = lev,
           colors = '#ff8243')
axes[0].set_title('Линии одного цвета')

#  Создаем массив RGB цветов каждого уровня:
color_line = np.zeros((10, 3))
color_line[:, 1:] = 0.1
color_line[:, 0] = np.linspace(0, 1, 10)

#  Контуры разного цвета:
axes[1].contour(x, y, E, levels = lev,
           colors = color_line)
axes[1].set_title('Линии разного цвета')

fig.set_figwidth(12)    #  ширина и
fig.set_figheight(6)    #  высота "Figure"

plt.show()
```

```python
#Если области между контурами будут заполнены разными цветами то мы добьемся того же самого результата. Метод contourf 
#выделяет контуры не линиями, а цветом областей между ними:

x, y = np.mgrid[-3*np.pi:3*np.pi:300j,
                -3*np.pi:3*np.pi:300j]
z = np.sinc(x) + np.cos(y)

fig, axes = plt.subplots(1, 2)

axes[0].contour(z)
axes[0].set_title('contour')

axes[1].contourf(z)
axes[1].set_title('contourf')

fig.set_figwidth(12)    #  ширина и
fig.set_figheight(6)    #  высота "Figure"

plt.show()
```
