## Pseudo Random and True Random.
Computers work on programs, and programs are definitive set of instructions. So it means there must be some algorithm to generate a random number as well.

If there is a program to generate random number it can be predicted, thus it is not truly random.

Random numbers generated through a generation algorithm are called pseudo random.

Can we make truly random numbers?

Yes. In order to generate a truly random number on our computers we need to get the random data from some outside source. This outside source is generally our keystrokes, mouse movements, data on network etc.

We do not need truly random numbers, unless its related to security (e.g. encryption keys) or the basis of application is the randomness (e.g. Digital roulette wheels).

In this tutorial we will be using pseudo random numbers.

# Случайные распределения

```python
"""
Нормальное распределение с центром в нуле и единичной дисперсией
"""

normal = random.normal(size=100)
normal
```

```python
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(12,12))
plt.grid()
hist = plt.hist(normal, bins=20)
```

```python
normal = random.normal(size=10000)
```

```python
plt.figure(figsize=(12,12))
plt.grid()
hist = plt.hist(normal, bins=200)
```

```python
"""
А что, если мы хотим сдвинуть центр распределения в точку mu?
"""
mu=10
normal = random.normal(loc=mu, size=10000)
plt.figure(figsize=(12,12))
plt.grid()
hist = plt.hist(normal, bins=200)
```

```python
mu=-100
normal = random.normal(loc=mu, size=10000)
plt.figure(figsize=(12,12))
plt.grid()
hist = plt.hist(normal, bins=200)
```

```python
"""
А если хотим изменить стандартное отклонение?
"""

sigma=10
normal = random.normal(scale=sigma, size=10000)
plt.figure(figsize=(12,12))
plt.grid()
hist = plt.hist(normal, bins=200)
```

```python
sigma=1000
normal = random.normal(scale=sigma, size=10000)
plt.figure(figsize=(12,12))
plt.grid()
hist = plt.hist(normal, bins=200)
```

```python
"""
Комбинация аргументов позволит нам делать и то, и другое
"""
mu = 15_000
sigma = 1000
normal = random.normal(loc=mu, scale=sigma, size=10000)
plt.figure(figsize=(12,12))
plt.grid()
hist = plt.hist(normal, bins=200)
```

```python
"""
Альтернатива такому подходу - задать распределение руками
"""
mu = 10
sigma = 3

normal = random.normal(size=10000)*sigma + mu

plt.figure(figsize=(12,12))
plt.grid()
hist = plt.hist(normal, bins=200)
```

```python
mu = np.array([10,100,1000]).reshape(-1,1)
sigma = np.array([3,30,300]).reshape(-1,1)

normal = random.normal(loc=mu, scale=sigma, size=(3,10000))
plt.figure(figsize=(12,12))
plt.grid()
hist = plt.hist(normal[0,:], bins=200)
```

```python
plt.figure(figsize=(12,12))
plt.grid()
hist = plt.hist(normal[1,:], bins=200)
```

```python
plt.figure(figsize=(12,12))
plt.grid()
hist = plt.hist(normal[2,:], bins=200)
```

```python
"""
Равномерное распределение на отрезке от 0 до 1
"""

uniform = random.uniform(size=10000)

plt.figure(figsize=(12,12))
plt.grid()
hist = plt.hist(uniform, bins=200)
```

```python
lambd = 100
exponential = random.exponential(lambd, size=1000000)

plt.figure(figsize=(12,12))
plt.grid()
hist = plt.hist(exponential, bins=200)
```

```python
"""
Распределение Парето
(Закон Ципфа из лингвистики)
"""

k = 1000
pareto = random.pareto(k, size=1000000)

plt.figure(figsize=(12,12))
plt.grid()
hist = plt.hist(pareto, bins=200)
```

## Приведем вполне реалистичный пример использования вероятностных распределений

### Пусть перед вами стоит задача максимизации функции f(x,y,z), о которой Вы ничего не знаете кроме того, что максимум этой функции лежит в пределах |x| < 10, |y| < 10, |z| < 10

### Если Вы ничего не знаете о функции, то подбор оптимальных значений можно осуществить перебором
### Однако есть ограничения на число комбинаций аргументов, которые Вы можете перебрать

<img src="random_grid.PNG">

```python
"""
Попробуем решить такую задачу
"""
f = lambda x,y,z: (-x**2 + 100)*(-y**2 + 6*y + 100)*(-z**2 - 4*z + 100)
```

```python
f(1,2,3)
```

```python
f(5,4,3)
```

```python
"""
Допустим, мы можем перебрать 27 вариантов
"""
values = []
for x in np.arange(-10, 10, 20/3):
    for y in np.arange(-10, 10, 20/3):
        for z in np.arange(-10, 10, 20/3):
            values.append(f(x,y,z))
values
```

```python
max(values)
```

```python
"""
Теперь попробуем перебрать варианты из случайных распределений
"""

x_values = np.random.uniform(-10,10,3)
y_values = np.random.uniform(-10,10,3)
z_values = np.random.uniform(-10,10,3)
```

```python
values = []
for x in x_values:
    for y in y_values:
        for z in z_values:
            values.append(f(x,y,z))
values
```

```python
max(values)
```

```python
"""
Если Вы все же что-то знаете про функцию f, Вы можете сделать дополнительные предположения. 
Скажем, что семплировать нужно из нормального распределения
"""

x_values = np.random.normal(loc=0, scale=3, size=3)
y_values = np.random.normal(loc=0, scale=3, size=3)
z_values = np.random.normal(loc=0, scale=3, size=3)
```

```python
values = []
for x in x_values:
    for y in y_values:
        for z in z_values:
            values.append(f(x,y,z))
values
```

```python
max(values)
```
